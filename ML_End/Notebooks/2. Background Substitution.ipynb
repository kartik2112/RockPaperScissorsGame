# Main Execution Background Substition Loop

```python
light_green = np.array((40, 40,40),np.uint8)
dark_green = np.array((70, 255,255),np.uint8)

bg_filenames = os.listdir("../datasets/backgrounds_augmented/")

for folder in ['paper/', 'rock/', 'scissors/']:
    filepath = "../datasets/kaggle_dataset/"+folder
    os.system("rm ../datasets/generated_dataset/" + folder+"*")
    for filename in tqdm(os.listdir(filepath)):
        for i in range(10):
            try:
                img = cv2.imread(filepath+filename)
    #             print(img)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
                mask = 255-(cv2.inRange(img, light_green, dark_green))
                result = cv2.bitwise_and(img, img, mask=mask)
                result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)
                bg_f = random.choice(bg_filenames)
                bg = cv2.cvtColor(cv2.imread("../datasets/backgrounds_augmented/"+bg_f), cv2.COLOR_BGR2RGB)
                bg_mask1 = cv2.bitwise_and(bg, bg, mask=(255-mask))
                new_img = cv2.max(bg_mask1, result)
                new_img = cv2.cvtColor(new_img, cv2.COLOR_RGB2BGR)
                cv2.imwrite("../datasets/generated_dataset/"+ folder + "".join(filename.split(".")[:-1])+"_"+str(i)+".png", new_img)
#                 imshow(new_img)
                del img, new_img, mask, bg
#                 time.sleep(1) 
            except:
                print(f"Read failed for image: {filename}")
#                 raise
```
