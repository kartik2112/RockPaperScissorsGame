```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import os
from tqdm import tqdm
from skimage.io import imshow
import time
```

```python
img = cv2.imread("../datasets/kaggle_dataset/paper/0cb6cVL8pkfi4wF6.png")
```

```python
def overlayHand(handImg, bgImg, lowColor, highColor, chosenSide):
    
    handImg = cv2.cvtColor(handImg, cv2.COLOR_RGB2HSV)
#     bgImg = cv2.cvtColor(bgImg, cv2.COLOR_RGB2HSV)
    
    # Compare bg and hand image dimensions
    handHeight, handWidth, _ = handImg.shape
    bgHeight, bgWidth, _ = bg.shape
    handAR = handWidth/handHeight
    
    widDiff = max(handWidth - bgWidth, 0)
    htDiff = max(handHeight - bgHeight, 0)
    
    # Resize hand image
    if (widDiff or htDiff):
        if widDiff >= htDiff:
            handImg = cv2.resize(handImg, (bgWidth, int(bgWidth / handAR)), interpolation = cv2.INTER_AREA)
            handWidth, handHeight, _ = handImg.shape
        else:
            handImg = cv2.resize(handImg, (int(bgHeight * handAR), bgHeight), interpolation = cv2.INTER_AREA)
            handWidth, handHeight, _ = handImg.shape
    
    # Generate hand mask and extract
    handMask = 255-(cv2.inRange(handImg, lowColor, highColor))
    handMasked = cv2.bitwise_and(handImg, handImg, mask=handMask)
    handMasked = cv2.cvtColor(handMasked, cv2.COLOR_HSV2RGB)
    
    assert handMask.shape == handImg.shape[:2]
    assert handMasked.shape == handImg.shape
    
    # Generate negative bg mask and subtract
    bgMask = np.zeros((bgHeight, bgWidth), dtype=np.uint8)
    # Add centered hand mask
    if chosenSide == 'bottom':
        startY = bgHeight - handHeight
        startX = bgWidth // 2 - handWidth // 2
    elif chosenSide == 'right':
        startY = bgHeight // 2 - handHeight // 2
        startX = bgWidth - handWidth
    
    bgMask[startY:startY + handHeight, startX:startX + handWidth] = handMask
    bgMask = 255 - bgMask
    assert bgMask.shape == bg.shape[:2]
    
    bgMasked = cv2.bitwise_and(bg, bg, mask=bgMask)
#     bgMasked = cv2.cvtColor(bgMasked, cv2.COLOR_HSV2RGB)
    
    # Extend handMasked Image
    handMaskedExt = np.zeros(bg.shape, dtype=np.uint8)
    handMaskedExt[startY:startY + handHeight, startX:startX + handWidth] = handMasked
    
    return cv2.max(bgMasked, handMaskedExt)
```

```python
light_green = np.array((40, 40,40),np.uint8)
dark_green = np.array((70, 255,255),np.uint8)

bg_filenames = os.listdir("../datasets/backgrounds_augmented/")

for folder in ['paper/', 'rock/', 'scissors/']:
    filepath = "../datasets/kaggle_dataset/"+folder
    os.system("rm ../datasets/generated_dataset/" + folder+"*")
    for filename in tqdm(os.listdir(filepath)):
        for i in range(10):
            try:
                img = cv2.imread(filepath+filename)
    #             print(img)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
                mask = 255-(cv2.inRange(img, light_green, dark_green))
                result = cv2.bitwise_and(img, img, mask=mask)
                result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)
                bg_f = random.choice(bg_filenames)
                bg = cv2.cvtColor(cv2.imread("../datasets/backgrounds_augmented/"+bg_f), cv2.COLOR_BGR2RGB)
                bg_mask1 = cv2.bitwise_and(bg, bg, mask=(255-mask))
                new_img = cv2.max(bg_mask1, result)
                new_img = cv2.cvtColor(new_img, cv2.COLOR_RGB2BGR)
                cv2.imwrite("../datasets/generated_dataset/"+ folder + "".join(filename.split(".")[:-1])+"_"+str(i)+".png", new_img)
#                 imshow(new_img)
                del img, new_img, mask, bg
#                 time.sleep(1) 
            except:
                print(f"Read failed for image: {filename}")
#                 raise
```
